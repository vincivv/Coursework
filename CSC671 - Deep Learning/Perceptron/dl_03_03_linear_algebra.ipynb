# Weighted Sum in Python

$$
z = b + \sum_{j=1}^m x_j w_j
$$

```python
b = 0.
x = [1.3, 2.5]
w = [3.6, 4.7]

output = b
for x_j, w_j in zip(x, w):
    output += x_j * w_j
    
print(output)
```

\begin{align*}
z &= b + \sum_{j=1}^m x_j w_j \\
&= b + \mathbf{xw}
\end{align*}


```python
import torch
b = torch.tensor(0.)
x = torch.tensor([1.3, 2.5])
w = torch.tensor([3.6, 4.7])

x.dot(w) + b
```

# Speed Improvement of PyTorch vs Plain Python for Vector-Vector Dot Product

```python
def plain_python(x, w, b):
    output = b
    for x_j, w_j in zip(x, w):
        output += x_j * w_j 
    return(output)
```

```python
import random
random.seed(737)

b = 0.
x = [random.random() for _ in range(1000)]
w = [random.random() for _ in range(1000)]
```

```python
result1 = %timeit -o plain_python(x, w, b) 
```

```python
def pytorch_dot_product(x, w, b):
    return (x.dot(w) + b)
```

```python
t_b = torch.tensor(b)
t_x = torch.tensor(x)
t_w = torch.tensor(w)
```

```python
result2 = %timeit -o pytorch_dot_product(t_x, t_w, t_b)
```

```python
print(f'speedup = {result1.average / result2.average}')
```

# Speed Improvement of Matrix-Vector Product Over For Loop

```python
import random
random.seed(737)

b = 0.
X = [[random.random() for _ in range(1000)]
     for _ in range(500)]
w = [random.random() for _ in range(1000)]
```

```python
def plain_matrix_vector_multiply(X, w, b):
    outputs = []
    for x in X:
        output = b
        for x_j, w_j in zip(x, w):
            output += x_j * w_j
        outputs.append(output)
    return outputs
```

```python
res_plain_matrix = %timeit -o plain_matrix_vector_multiply(X, w, b)
```

```python
t_b = torch.tensor(b)
t_X = torch.tensor(X)
t_w = torch.tensor(w)
```

```python
res_pytorch_matrix = %timeit -o t_X.matmul(t_w) + t_b
```

```python
print(f'speedup = {res_plain_matrix.average / res_pytorch_matrix.average}')
```
